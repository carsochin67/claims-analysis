```python
## Question 1: Who are the top 5 billing providers by number of claims?
provider_col = "BillingProviderNPI"   # <- check actual column name

claims_per_provider = (
    df_header
    .groupby(provider_col)["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

top5_providers = claims_per_provider.head(5)
display(top5_providers)

```

This block of code identifies the top 5 billing providers by counting how many claims each provider submitted. It groups claims by BillingProviderNPI, counts the unique claims per provider, sorts them, and displays the top contributors. The results show that one provider (NPI 1821035601) billed 152 claims, far more than any other provider, making them the highest-volume biller in this dataset. This indicates that a small number of providers account for a large share of claim volume, which is important for understanding workload, billing patterns, and operational priorities.

```python
plt.figure()
plt.bar(top5_providers[provider_col].astype(str), top5_providers["claim_count"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 5 Billing Providers by Number of Claims")
plt.xlabel("Provider NPI")
plt.ylabel("Number of Claims")
plt.tight_layout()
plt.show()
```

This block of code creates a bar chart that visualizes the top five billing providers based on the number of claims they submitted. The purpose of the plot is to give a quick, easy-to-read comparison of provider claim volume, which helps highlight who is driving the majority of billing activity in this dataset. The chart shows that one provider (NPI 1821035601) billed far more claims than anyone else. This big gap suggests that this provider handles a much larger share of patient encounters or billing workload, and they likely play a major role in the overall claim volume for this facility.

```python
## Question 2:Payer Mix Analysis Top 5 primary payers by claim volume + % of total

payer_col = "PrimaryPayerName"   # <- check actual name

claims_per_payer = (
    df_header
    .groupby(payer_col)["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

total_claims = claims_per_payer["claim_count"].sum()
claims_per_payer["pct_of_total"] = claims_per_payer["claim_count"] / total_claims * 100

top5_payers = claims_per_payer.head(5)
display(top5_payers)
```

This block of code analyzes the payer mix by identifying the top five primary insurance payers based on how many claims they cover. The code groups all claims by the primary payer name, counts how many unique claims each payer is responsible for, and then calculates what percentage of the total claim volume each payer represents.The results show that Medicare accounts for the overwhelming majority of claims, which is about 62% of the entire dataset.

```python
plt.figure()
plt.pie(top5_payers["claim_count"],
        labels=top5_payers[payer_col],
        autopct="%1.1f%%")
plt.title("Top 5 Primary Payers by Claim Volume")
plt.show()
```

This pie chart visualizes the top five primary payers based on their share of total claims. The purpose of the chart is to show how the payer mix is distributed and whether one payer dominates the dataset. From the chart, its clear that Medicare makes up the vast majority of claims, which is about 71%, which means most of the patients seen here are covered by Medicare.

```python
## Question 3: Common Diagnoses Top 10 most frequent ICD-10 codes
dx_counts = (
    df_code
    .groupby("CodeValue")["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

top10_dx = dx_counts.head(10)
display(top10_dx)
```

This block of code identifies the top ten most common diagnosis codes in the dataset. It groups all diagnosis records by the code itself, counts how many unique claims each diagnosis appears on, sorts them from most to least frequent, and then returns the top ten. This mix suggests a strong presence of patients with chronic conditions like hypertension and hyperlipidemia, along with serious acute issues such as respiratory failure, anemia, and neurologic disorders.

```python
## Q4  Common Procedures Top 10 HCPCS procedure codes

proc_col = "HCPCS"  # or "CPT_HCPCS" depending on file

proc_counts = (
    df_line
    .groupby(proc_col)["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

top10_proc = proc_counts.head(10)
display(top10_proc)

plt.figure()
plt.bar(top10_proc[proc_col].astype(str), top10_proc["claim_count"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Procedure Codes by Number of Claims")
plt.xlabel("HCPCS Code")
plt.ylabel("Number of Claims")
plt.tight_layout()
plt.show()
```

This block of code identifies the top ten most common procedures based on HCPCS codes and visualizes them in a bar chart. It groups the LINE file by the HCPCS code, counts how many claims include each procedure, sorts them from most to least frequent, and then displays the top ten. From the results, the most frequent procedure by far is 99291, which is the code for critical care, first hour. This appears 68 times, making it the most common service in the dataset. Other top procedures include 99233 and 99213, which are inpatient and outpatient evaluation and management (E/M) visits.

```python
## Question 5: Service Location Analysis

pos_col = "PlaceOfService"  # check actual name

claims_per_pos = (
    df_header
    .groupby(pos_col)["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

display(claims_per_pos)

total_claims = claims_per_pos["claim_count"].sum()
claims_per_pos["pct_of_total"] = claims_per_pos["claim_count"] / total_claims * 100
display(claims_per_pos)
```

```python
## Part 3  Advanced Analysis with Joins Helper: merged Header + Line
hl = df_header.merge(df_line, on="ProspectiveClaimId", how="inner")
```

```python
## Question 6: Claims with High Service Line Counts

lines_summary = (
    hl
    .groupby("ProspectiveClaimId")
    .agg(
        num_lines=("LinePos", "count"),       # or another column thats 1 per service line
        total_charges=("Charges", "sum")      # check column name
    )
    .reset_index()
)

high_line_claims = lines_summary[lines_summary["num_lines"] >= 5]
display(high_line_claims.head())

# join back to header for provider + payer info
high_line_claims_detail = high_line_claims.merge(
    df_header,
    on="ProspectiveClaimId",
    how="left"
)

display(high_line_claims_detail.head())
```

This section of code looks for claims that have unusually high numbers of service lines. The results show example claims with 57 service lines each, with total charges ranging from the high hundreds to over a thousand dollars. Many of these high-line claims come from OB/GYN or university-associated clinics, which suggests these locations may be handling more complex visits requiring multiple services in one encounter.

```python
## Question 7: DiagnosisProcedure Combinations; Goal: What diagnosis codes commonly appear with CPT/HCPCS 99291?

# Filter to lines with the target procedure
target_code = "99291"
proc_col = "HCPCS"  # or "CPT_HCPCS"

lines_99291 = df_line[df_line[proc_col] == target_code]

# Join with CODE file to get diagnosis codes for those claims
claims_99291_dx = lines_99291[["ProspectiveClaimId"]].merge(
    df_code,
    on="ProspectiveClaimId",
    how="inner"
)

dx_counts_99291 = (
    claims_99291_dx
    .groupby("CodeValue")["ProspectiveClaimId"]
    .nunique()
    .reset_index(name="claim_count")
    .sort_values("claim_count", ascending=False)
)

top_dx_99291 = dx_counts_99291.head(10)
display(top_dx_99291)
```

This block of code looks at which diagnosis codes most commonly appear on claims that include the procedure code 99291, which is the CPT/HCPCS code for critical care, first hour. The results show that the diagnoses most frequently associated with critical care claims include things like acute respiratory failure (J96.01), sepsis (G93.5), hyperlipidemia (E78.5), heart failure (I61.9), and pneumonia (J10/I10).

```python
## Question 8: Charges by Payer

# total charges per claim
claim_charges = (
    hl
    .groupby("ProspectiveClaimId")["Charges"]   # check column name
    .sum()
    .reset_index(name="total_claim_charges")
)

# attach payer
claim_charges = claim_charges.merge(
    df_header[["ProspectiveClaimId", payer_col]],
    on="ProspectiveClaimId",
    how="left"
)

payer_charge_stats = (
    claim_charges
    .groupby(payer_col)
    .agg(
        total_charges=("total_claim_charges", "sum"),
        avg_charge_per_claim=("total_claim_charges", "mean"),
        num_claims=("ProspectiveClaimId", "nunique")
    )
    .reset_index()
    .sort_values("total_charges", ascending=False)
)

top10_payers_charges = payer_charge_stats.head(10)
display(top10_payers_charges)
```

This block of code looks at how much each payer is being billed overall, and it also calculates how many claims each payer has and the average charge per claim. It does this by first summing up all service-line charges for each claim in the LINE file. The results show that Medicare is by far the dominant payer in this dataset.

```python
## Part 4  Question 9: Creative Analysis

# reuse claim_charges from above
claim_pos = df_header[["ProspectiveClaimId", pos_col]]
claim_charges_pos = claim_charges.merge(claim_pos, on="ProspectiveClaimId", how="left")

pos_charge_stats = (
    claim_charges_pos
    .groupby(pos_col)["total_claim_charges"]
    .agg(["count", "mean", "sum"])
    .reset_index()
    .rename(columns={"count": "num_claims", "mean": "avg_charge", "sum": "total_charges"})
    .sort_values("avg_charge", ascending=False)
)

display(pos_charge_stats)

plt.figure()
plt.bar(pos_charge_stats[pos_col].astype(str), pos_charge_stats["avg_charge"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("Average Charge per Claim")
plt.title("Average Claim Charges by Place of Service")
plt.tight_layout()
plt.show()
```

This block of code explores how average claim charges differ by Place of Service (POS). The results show that Place of Service 22 has the highest average charge per claim at around $990. This POS category represents the large majority of claims in the dataset.
